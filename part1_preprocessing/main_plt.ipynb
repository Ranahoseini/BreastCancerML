```python
import matplotlib.pyplot as plt
```

```python
x = [1,1, 3, 4, 5, 6, 7, 8, 9]
y = [2, 2, 9, 3, 6, 7, 1, 4, 8]

plt.plot(x, y,marker = '*' ,color = "blue")
plt.title("Rana")
plt.xlabel("x")
plt.ylabel("Y")


plt.show()
```

```python
x= ["red", "yellow", "green", "gray"]
y= [3, 5, 8, 1]

plt.bar(x, y, color = "yellow")
plt.title("Rana")
plt.xlabel("color")
plt.ylabel("count")


plt.show()
```

```python
x= ["red", "yellow", "green", "gray"]
y= [3, 5, 8, 1]

plt.pie(y, labels = x, autopct = "%1.1f%%", colors = ["red", "yellow", "green", "gray"])
plt.show()
```
